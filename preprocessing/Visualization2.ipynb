```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Read csv file into a pandas dataframe

df = pd.read_csv("..\dataset\players_cleaned.csv")

# Take a look at the first few rows
print (df.head())
```
