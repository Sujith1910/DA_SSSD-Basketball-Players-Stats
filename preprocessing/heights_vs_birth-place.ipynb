```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../datasets/Players.csv")

# Take a look at the first few rows
print (df.head())
```

```python
bs1 = df.groupby(['birth_state']).count().reset_index().sort_values(by='height',ascending=False)
grouped_height = df['height'].groupby(df['birth_state']) 
mean_height = grouped_height.mean()

#printing the means value 
#print(grouped_height.mean()) 
#print(mean_height)
```
