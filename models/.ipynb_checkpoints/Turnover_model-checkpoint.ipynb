```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import linear_model
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../datasets/seasons_norm_dataset.csv")

grouped_by_year = df.groupby(['Year'])
stats_2017 = df[df['Year']==2017]
stats_2016 = grouped_by_year.get_group(2016)
```

```python
#Turnover calculation 

X = stats_2017[['MP','Age','FG','FT','ORB','AST','PF']]
y = stats_2017['TOV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=True)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
```

```python
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
y_pred=regr.predict(X_test)
```

```python
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_squared_error,r2_score

mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
x_for_test=[i+1 for i in range(len(y_test))]
plt.plot(x_for_test,y_pred,color='Red',label="Predicted")
plt.plot(x_for_test,y_test,color='Blue',label="Actual")
plt.xlabel('Player #')
plt.ylabel('TOV')
plt.title('TOV Model')
plt.legend()
plt.show()
```

```python
regr.score(X_test,y_test)
```
